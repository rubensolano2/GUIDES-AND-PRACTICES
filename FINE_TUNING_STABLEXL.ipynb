```python
from diffusers import DiffusionPipeline
import torch

model = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype= torch.float16,
)
pipe.to("cuda")
pipe.load_lora_weights("/content/drive/MyDrive/1. proyectos/trabajo fiver/proyectos/fine_tune_05_10_23/pytorch_lora_weights.safetensors", weight_name="pytorch_lora_weights.safetensors"
)
prompt = "POV driving in realistic urban and rural settings, captured in 4K resolution, showcasing the driver's perspective as they navigate through various traffic scenarios, obeying traffic rules and signs, with realistic reflections, shadows, and interactions with other road users. The image captures the essence of a typical driving experience, complete with the vivid colors of the environment, the textures of the road, and the dynamics of moving vehicles."  # Corregido 'promt' a 'prompt'

for seed in range(5):
  generator = torch.Generator("cuda").manual_seed(seed)
  image = pipe(prompt=prompt, generator=generator, num_inference_steps=25)  # Corregido 'promt' a 'prompt'
  image = image.images[0]
  image.save(f"/content/drive/MyDrive/1. proyectos/trabajo fiver/proyectos/fine_tune_05_10_23/results/{seed}.png")
```
