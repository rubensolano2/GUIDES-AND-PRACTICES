### ENCODER

Este codificador toma una oración de entrada y la convierte en una forma que la máquina pueda entender, mediante "tokens". Luego, estos tokens se pasan a través de una red LSTM (Long Short-Term Memory), que es un tipo de red neuronal recurrente. Al final, el codificador devuelve un "estado oculto", que es un conjunto de valores que representa la información extraída de la oración de entrada. Este estado se puede utilizar en pasos posteriores del modelo.

Detalles importantes:

1. **`bidirectional`:** Este parámetro indica si la LSTM es unidireccional o bidireccional. En una LSTM unidireccional, la información solo fluye de izquierda a derecha a través de la secuencia. En una LSTM bidireccional, hay dos LSTMs: una procesa la secuencia de izquierda a derecha y la otra de derecha a izquierda. Los resultados de ambas se suelen combinar en cada paso de tiempo. En tu código, `bidirectional=False` significa que estás usando una LSTM unidireccional.

2. **`batch_first`:** Normalmente, las dimensiones de entrada a una LSTM en PyTorch son `(seq_len, batch, input_size)`. Sin embargo, si `batch_first=True`, puedes enviar los datos de entrada con la forma `(batch, seq_len, input_size)`. Esto a menudo simplifica el código, especialmente si estás trabajando con otros módulos o capas que esperan esta ordenación de dimensiones. En tu caso, `batch_first=True` implica que debes organizar tus datos de entrada con la dimensión del lote (`batch`) como la primera dimensión.

Ambos parámetros son configurables y su elección dependerá de las necesidades de tu tarea específica y de cómo estás procesando y preparando tus datos.


¿Cuando usar bidireccional?

El uso de una LSTM bidireccional generalmente depende del tipo de problema y de los datos que estés manejando. Aquí hay algunas pautas sobre cuándo podrías querer utilizar una:

1. **Contexto de ambas direcciones**: Si tanto el contexto pasado como el futuro son importantes para tu tarea, una LSTM bidireccional es una buena opción. Por ejemplo, en tareas de etiquetado de secuencias como el reconocimiento de entidades nombradas, el contexto futuro puede ser tan importante como el contexto pasado.

2. **Datos Simétricos**: Si tus datos no tienen una dirección temporal inherente (por ejemplo, la estructura de una oración en un lenguaje natural), una LSTM bidireccional podría captar mejor las dependencias.

3. **Mejor rendimiento**: En muchas tareas, simplemente se ha encontrado que las LSTM bidireccionales tienen un rendimiento superior al de las unidireccionales, aunque a costa de ser más complejas y requerir más recursos computacionales.

4. **No hay restricciones temporales**: En algunos casos, como el procesamiento en tiempo real o las series temporales en las que el futuro aún no está disponible, no podrás usar una LSTM bidireccional.

5. **Recursos Computacionales**: Ten en cuenta que una LSTM bidireccional tiene el doble de parámetros que una unidireccional, lo que aumenta la complejidad y el tiempo de entrenamiento.

En resumen, si tu tarea se beneficia del contexto tanto pasado como futuro y no tienes restricciones severas de tiempo o recursos, una LSTM bidireccional podría ser una buena opción.

Relleno en el vector (padding):

El "relleno" (padding) es una técnica común en el procesamiento de secuencias, como el texto, donde las secuencias pueden tener diferentes longitudes pero se necesita un tensor de entrada de forma fija para el modelo. Por ejemplo, si tienes secuencias de longitud 5, 6 y 7, podrías rellenarlas hasta una longitud de 10, añadiendo ceros (u otro valor específico) para que todas tengan la misma longitud.

Ahora, este relleno puede ser añadido ya sea al inicio (izquierda) o al final (derecha) de la secuencia. Algunos modelos o bibliotecas pueden tener una preferencia por uno de los dos métodos. En el código que has mostrado, se verifica si el relleno está en la izquierda (`self.args.left_pad_source`) y, si es así, lo mueve a la derecha.

La razón para mover el relleno de izquierda a derecha podría ser una cuestión de consistencia o rendimiento, dependiendo de cómo se haya implementado el resto del modelo o la biblioteca que se esté utilizando. No hay un beneficio intrínseco en tener el relleno en un lado u otro; más bien, es importante que el modelo sepa dónde encontrarlo para que pueda manejar las secuencias de manera efectiva.

##DECODER

Nuestro decodificador predecirá la siguiente palabra, condicionado por el estado oculto final del codificador y una representación incrustada de la palabra objetivo anterior, lo cual a veces se llama "forzado del profesor". Más específicamente, utilizaremos un torch.nn.LSTM para producir una secuencia de estados ocultos que proyectaremos al tamaño del vocabulario de salida para predecir cada palabra objetivo.

```python
import torch
from fairseq.models import FairseqDecoder
```

```python


class SimpleLSTMDecoder(FairseqDecoder):

    def __init__(
        self, dictionary, encoder_hidden_dim=128, embed_dim=128, hidden_dim=128,
        dropout=0.1,
    ):
        super().__init__(dictionary)

        # Nuestro decodificador incrustará las entradas antes de alimentarlas al LSTM.
        self.embed_tokens = nn.Embedding(
            num_embeddings=len(dictionary),
            embedding_dim=embed_dim,
            padding_idx=dictionary.pad(),
        )
        self.dropout = nn.Dropout(p=dropout)

        # Utilizaremos un LSTM de una sola capa y unidireccional por simplicidad.
        self.lstm = nn.LSTM(
            # Para la primera capa concatenaremos el estado oculto final del codificador
            # con los tokens objetivo incrustados.
            input_size=encoder_hidden_dim + embed_dim,
            hidden_size=hidden_dim,
            num_layers=1,
            bidirectional=False,
        )

        # Definir la proyección de salida.
        self.output_projection = nn.Linear(hidden_dim, len(dictionary))

    # Durante el entrenamiento, se espera que los decodificadores tomen toda la secuencia objetivo
    # (desplazada a la derecha por una posición) y produzcan logits sobre el vocabulario.
    # El tensor *prev_output_tokens* comienza con el símbolo de fin de oración,
    # ``dictionary.eos()``, seguido de la secuencia objetivo.
    def forward(self, prev_output_tokens, encoder_out):
        """
        Args:
            prev_output_tokens (LongTensor): salidas anteriores del decodificador de forma
                `(batch, tgt_len)`, para el forzado del profesor.
            encoder_out (Tensor, opcional): salida del codificador, usada para
                la atención del lado del codificador.

        Devoluciones:
            tupla:
                - la última salida de la capa del decodificador de forma
                  `(batch, tgt_len, vocab)`
                - los últimos pesos de atención de la capa del decodificador de forma
                  `(batch, tgt_len, src_len)`
        """
        bsz, tgt_len = prev_output_tokens.size()

        # Extraer el estado oculto final del codificador.
        final_encoder_hidden = encoder_out['final_hidden']

        # Incrustar la secuencia objetivo, que ha sido desplazada a la derecha por una
        # posición y ahora comienza con el símbolo de fin de oración.
        x = self.embed_tokens(prev_output_tokens)

        # Aplicar dropout.
        x = self.dropout(x)

        # Concatenar el estado oculto final del codificador a *cada* token objetivo incrustado.
        x = torch.cat(
            [x, final_encoder_hidden.unsqueeze(1).expand(bsz, tgt_len, -1)],
            dim=2,
        )

        # Usar objetos PackedSequence en el decodificador es más difícil que en el
        # codificador, ya que los objetivos no están ordenados en orden descendente de longitud,
        # que es un requisito de ``pack_padded_sequence()``. En lugar de eso, alimentaremos directamente a nn.LSTM.
        initial_state = (
            final_encoder_hidden.unsqueeze(0),  # hidden
            torch.zeros_like(final_encoder_hidden).unsqueeze(0),  # cell
        )
        output, _ = self.lstm(
            x.transpose(0, 1),  # convertir a forma `(tgt_len, bsz, dim)`
            initial_state,
        )
        x = output.transpose(0, 1)  # convertir a forma `(bsz, tgt_len, hidden)`

        # Proyectar las salidas al tamaño del vocabulario.
        x = self.output_projection(x)

        # Devolver los logits y ``None`` para los pesos de atención
        return x, None
```

El decodificador es conceptualmente similar al codificador, pero con algunas diferencias clave. Mientras que el codificador procesa la secuencia de entrada para crear un estado oculto final (o un conjunto de estados ocultos), el decodificador utiliza ese estado para generar una secuencia de salida.

Las similitudes incluyen el uso de una capa de incrustación (embedding), una capa LSTM y dropout. Sin embargo, el decodificador también tiene componentes adicionales, como la proyección de salida que convierte los estados ocultos en logits sobre el vocabulario, y puede tener mecanismos de atención para centrarse en diferentes partes de la entrada durante la decodificación. Además, la entrada al LSTM del decodificador incluye información del estado oculto final del codificador, lo que permite al decodificador tener en cuenta todo el contexto de la entrada al generar la secuencia de salida.

# 2. Configurando el modelo

El proceso de "registro" en este contexto se refiere a la acción de hacer que el modelo sea reconocible y utilizable por la biblioteca fairseq. Al registrar el modelo, se le asigna un nombre específico ('simple_lstm' en este caso), que luego se puede utilizar para referirse al modelo en otros contextos dentro de fairseq, como sus herramientas de línea de comando.


Para hacer esto, se utiliza el decorador register_model(). Este es un patrón en Python que modifica o extiende la funcionalidad de una función o clase, en este caso, indicando a fairseq que este modelo en particular debería ser añadido a su lista de modelos disponibles.

En términos de implementación, se espera que todos los modelos registrados sigan una interfaz base llamada BaseFairseqModel. Para modelos que son de tipo secuencia a secuencia (es decir, tienen un codificador y un decodificador), puedes optar por implementar la interfaz FairseqEncoderDecoderModel.

Finalmente, para completar el registro, se crea una "clase contenedora" que implementa la interfaz antes mencionada. Esta clase actúa como un puente entre tu implementación específica del modelo y las capacidades genéricas proporcionadas por fairseq.

```python
from fairseq.models import FairseqEncoderDecoderModel, register_model
```

```python


# Nota: el "decorador" register_model debe preceder inmediatamente a
# la definición de la clase del modelo.

@register_model('simple_lstm')#registra el modelo personallizdo con el nombre simple_ltsm usando un decorador
class SimpleLSTMModel(FairseqEncoderDecoderModel):#define una clase llamada SimpleLSTMModel que hereda de FairseqEnconderDecoderModel

    @staticmethod#Inica que la siguiente funión es un método estático de la clase
    def add_args(parser):
        # Los modelos pueden sobrescribir este método para añadir nuevos argumentos de línea de comando.
        # Aquí añadiremos algunos para configurar el dropout y la dimensionalidad.
        parser.add_argument(#añade argumentos específicos al analizador de línea de comandos para el encoder y el decoder
            '--encoder-embed-dim', type=int, metavar='N',
            help='dimensionalidad de las incrustaciones del codificador',
        )
        parser.add_argument(
            '--encoder-hidden-dim', type=int, metavar='N',
            help='dimensionality of the encoder hidden state',
        )
        parser.add_argument(
            '--encoder-dropout', type=float, default=0.1,
            help='encoder dropout probability',
        )
        parser.add_argument(
            '--decoder-embed-dim', type=int, metavar='N',
            help='dimensionality of the decoder embeddings',
        )
        parser.add_argument(
            '--decoder-hidden-dim', type=int, metavar='N',
            help='dimensionality of the decoder hidden state',
        )
        parser.add_argument(
            '--decoder-dropout', type=float, default=0.1,
            help='decoder dropout probability',
        )


    @classmethod#la siguiente función es un método de clase
    def build_model(cls, args, task):#define cómo construir el modelo. Toma argumentos de la línea de comandos y una tarea como entrada
        # Fairseq inicializa los modelos mediante esta función.
        # Esto ofrece más flexibilidad.
        encoder = SimpleLSTMEncoder(#inicia el encoder con los argumentos especificados. LA case SimpleLSTMEncoder no está definida pero está implementada en otro lugar
            args=args,
            dictionary=task.source_dictionary,
            embed_dim=args.encoder_embed_dim,
            hidden_dim=args.encoder_hidden_dim,
            dropout=args.encoder_dropout,
        )
        decoder = SimpleLSTMDecoder( #inicia en decoder con los arguments como el encoder
            dictionary=task.target_dictionary,
            encoder_hidden_dim=args.encoder_hidden_dim,
            embed_dim=args.decoder_embed_dim,
            hidden_dim=args.decoder_hidden_dim,
            dropout=args.decoder_dropout,
        )
        model = SimpleLSTMModel(encoder, decoder) #Crea una instancia de SimpleLSTMModel utilizando encoder y dcoder iniciados

        # Imprimir la arquitectura del modelo.
        print(model) #impre la arquitectura para depurar y validar

        return model

    # Podríamos sobrescribir ``forward()`` para más control, pero no es necesario.
    # Podemos heredar la implementación por defecto de FairseqEncoderDecoderModel.
    #
    # def forward(self, src_tokens, src_lengths, prev_output_tokens):
    #     encoder_out = self.encoder(src_tokens, src_lengths)
    #     decoder_out = self.decoder(prev_output_tokens, encoder_out)
    #     return decoder_out
```

Método estático:

método que pertenece a una clase en lugar de a cualquier instancia particular de la clase. Esto significa que puedes llamar al método estático sin crear una instancia de la clase. Se accede directamente desde la clase y no puede acceder o modificar los atributos o métodos específicos de la clase o de sus instancias.

Los métodos estáticos se definen usando la anotación @staticmethod en Python.

Ventajas de usar métodos estáticos:

mantener métodos que son relevantes para una clase dentro de la clase, aunque no interactúen con atributos de instancia de la clase.Mejora en rendimiento: No es necesario crear una instancia de la clase para utilizar el método, lo cual puede ahorrar memoria y tiempo de cómputo.Facilita el mantenimiento y la legibilidad: Al agrupar funcionalidades que están estrechamente relacionadas en una sola clase, el código es más fácil de entender y mantener.

Explicación mas simple:

Un método estático es una función que pertenece a una clase, pero no necesita acceder a ninguna información específica de un objeto de esa clase para hacer su trabajo. Se usa directamente desde la clase y no desde una instancia de la clase. Es útil para agrupar funciones que tienen relación con la clase, pero que no necesitan saber nada sobre los detalles específicos de un objeto de esa clase.

método de clase:

Un método de clase es un tipo de método que se vincula a la clase y no a una instancia de la clase. En Python, se define usando el decorador @classmethod. A diferencia de un método estático, un método de clase toma un primer argumento especial llamado cls, que representa la clase en sí, en lugar de una instancia de la clase.

Esto significa que puede acceder y modificar atributos de clase, que son compartidos por todas las instancias de esa clase. No puede acceder a atributos específicos de una instancia, pero sí puede afectar el estado de la clase como un todo.

Por ejemplo, podrías tener un método de clase que configure una tasa de interés para una clase CuentaBancaria, y ese cambio se reflejaría en todas las instancias de CuentaBancaria.

##Registro a la arquitectura

Finalmente, definamos una arquitectura con nombre con la configuración para nuestro modelo. Esto se hace con el decorador de función `register_model_architecture()`. A partir de ese momento, esta arquitectura con nombre se puede usar con el argumento de línea de comando `--arch`, por ejemplo, `--arch tutorial_simple_lstm`:

```python
from fairseq.models import register_model_architecture

# Se utiliza el decorador "register_model_architecture" para registrar la arquitectura.
# El primer argumento debe ser el nombre del modelo registrado anteriormente ('simple_lstm').
# La función que se registra aquí toma un solo argumento *args* y lo modifica en su lugar para que coincida con la arquitectura deseada.
@register_model_architecture('simple_lstm', 'tutorial_simple_lstm')
def tutorial_simple_lstm(args):
    # Usamos "getattr()" para dar prioridad a los argumentos que se dan explícitamente en la línea de comandos.
    # De esta forma, los valores predeterminados definidos aquí solo se utilizan cuando no se ha especificado otro valor.
    args.encoder_embed_dim = getattr(args, 'encoder_embed_dim', 256)
    args.encoder_hidden_dim = getattr(args, 'encoder_hidden_dim', 256)
    args.decoder_embed_dim = getattr(args, 'decoder_embed_dim', 256)
    args.decoder_hidden_dim = getattr(args, 'decoder_hidden_dim', 256)
```
